```python
# reversing a list in python

def listReverse(nums:list):
    n=len(nums)
    if(n<=1):
        return nums
    new_nums=nums[::-1]
    return new_nums

print(listReverse([10,20,30]))    
print(listReverse(['dsa','with','py']))
```

```python
# soltuion 2
def ReverseList(nums:list):
    n=len(nums)
    if(n<=1):
        return nums
    for i in range(n):
        nums.insert(i,nums.pop())
    return nums
print(ReverseList([10,20,30,40,50]))
```

```python
# solution 4
def ReverseList(nums:list):
    n=len(nums)
    s=0
    e=n-1
    while s<e:
        nums[s],nums[e]=nums[e],nums[s]
        s+=1
        e-=1
    return nums

print(ReverseList([10,20,30,40,50]))
print(ReverseList([10,20,30,40]))

# time complexity : O(n)
```
