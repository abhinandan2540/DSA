```python

def num_bit(n:int):
    while n>0:
        if(n%2!=0):
            print(1,end="")
        else:
            print(0,end="")
        n=n//2
    

num_bit(19)
  
```

```python
# returning bits of a number

def num_bit_1(num:int):
    num_bits=[]
    while num>0:
        if(num%2!=0):
            num_bits.append('1')
        else:
            num_bits.append('0')
        num=num//2
    b1=num_bits[::-1]
    b1_bits=''.join(b1)
    return b1_bits

print(num_bit_1(19))
```

```python
# using this func to calcuate the power of any number
def n_power(x:int, n:int):
    temp=1
    while n>0:
        if(n%2!=0):
            temp*=x
        x*=x
        n=n//2
    return temp

print(n_power(3,5))

# time complexity : O(logn)
```
