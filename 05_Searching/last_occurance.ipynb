```python
# unlike the 1st occurance
# for the last occurance

def lastOccurance(arr:list, k:int):
    low=0
    high=len(arr)-1
    while (low<=high):
        mid=(low+high)//2
        if(arr[mid]>k):
            high=mid-1
        elif (arr[mid]<k):
            low=mid+1
        else:
            if(mid==len(arr)-1 or arr[mid]!=arr[mid+1]):
                return mid
            else:
                low=mid+1
    return -1


print(lastOccurance([1,2,2,2,3,3,3,3,4],3))
# print(lastOccurance([1, 1, 1, 1, 1, 0, 0, 0],1))


# time complexity : O(logn)
```
