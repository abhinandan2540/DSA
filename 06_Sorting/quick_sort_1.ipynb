quick sort using *lomuto's algorithm*

```python
def lomuto_parititon(arr: list, l:int, h:int):
    pivot = arr[h] # taking the last number as pivot
    i = l-1
    for j in range(l,h):
        if (arr[j] <= pivot):
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i+1], arr[h] = arr[h], arr[i+1]
    return i+1


def q_sort(arr: list, l: int, h: int):
    if (l < h):
        p = lomuto_parititon(arr, l, h)
        q_sort(arr, l, p-1)
        q_sort(arr, p+1, h)
    return arr

a = [5, 13, 6,9,12,8,11]
print(q_sort(a,0,6))

```
