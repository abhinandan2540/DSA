```python
def union(a:list, b:list):
    res=[]
    m=len(a)
    n=len(b)
    i=0
    j=0
    while(i<m and j<n):
        if (a[i]<=b[j]):
            if(a[i] not in res):
                res.append(a[i])
            i+=1
        else:
            if (b[j] not in res):
                res.append(b[j])
            j+=1

    while(i<m):
        if (a[i] not in res):
            res.append(a[i])
        i+=1

    while(j<n):
        if(b[j] not in res):
            res.append(b[j])
        j+=1

    return res

# a=[3,5,8]
# b=[2,8,9,10,15]
a=[2,20,20,40]
b=[1,10,20]
print(union(a,b))

# time complexity : O(m+n)
```
