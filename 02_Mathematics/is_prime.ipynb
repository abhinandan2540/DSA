```python
# finding prime 
# optimized solution
# solution 1
def is_prime(n:int):
    isPrime=True
    if (n==1):
        isPrime=False
    i=2
    while (i*i<=n):
        if(n%i==0):
            isPrime=False
        i+=1
    return isPrime

print(is_prime(7))
print(is_prime(65))

# time complexity : O(n^1/2)
# space complexity : O(1) 
```

```python
# solution 3
def is_prime_2(n:int):
    isPrime=True
    if (n==1):
        isPrime=False
    if (n==2 or n==3):
        return isPrime
    if (n%2==0 or n%3==0):
        isPrime=False
    i=5
    while (i*i<=n):
        if(n%i==0) or (n%(i+2)==0):
            isPrime=False
        i+=6
    return isPrime

print(is_prime_2(67))
print(is_prime_2(2))
print(is_prime_2(4))
    
# time complexity :  O(n^1/2)
```
