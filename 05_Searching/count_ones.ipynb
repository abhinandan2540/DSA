```python

# finding the total counting of ones
# using the first occurance and last occurance method

def firstOccurance(arr:list):
    low = 0
    high = len(arr)-1
    while (low <= high):
        mid = (low+high)//2
        if(arr[mid]>1):
            high=mid-1
        elif (arr[mid]<1):
            low=mid+1
        else:
            if(mid==0 or arr[mid]!=arr[mid-1]):
                return mid
            else:
                high=mid-1
    return -1

def lastOccurance(arr:list):
    low = 0
    high = len(arr)-1
    while (low <= high):
        mid = (low+high)//2
        if(arr[mid]>1):
            high=mid-1
        elif (arr[mid]<1):
            low=mid+1
        else:
            if(mid==len(arr)-1 or arr[mid]!=arr[mid+1]):
                return mid
            else:
                low=mid+1
    return -1

def count_ones(arr:list):
    first_occurance=firstOccurance(arr)
    if first_occurance==-1:
        return 0
    else:
        return lastOccurance(arr)-first_occurance+1

print(count_ones([0,0,0,1,1,1,1]))
# print(count_ones([1,0]))
# print(firstOccurance([1 ,1 ,1, 1, 1, 0, 0, 0]))
# print(lastOccurance([1 ,1 ,1, 1, 1, 0, 0, 0]))
# print(count_ones([1 ,1 ,1, 1, 1, 0, 0, 0]))

# time complexity : O(logn)
                
```

```python
# there's another method for finding counting of 1's

def count_occurance(arr:list):
    n=len(arr)
    low=0
    high=n-1
    while(low<=high):

        mid=(low+high)//2
        if((arr[mid]==n-1 or arr[mid+1]==0) and arr[mid]==1):
            return mid+1
            # for left 
        elif (arr[mid]==0):
            high=mid-1
            # for right
        elif(arr[mid]==1):
            low=mid+1
    return 0

arr=[1,1,1,1,0,0,0]
print(count_occurance(arr))        
```
