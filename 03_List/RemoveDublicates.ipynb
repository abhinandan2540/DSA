```python
# removing dublicates from an array

# solution 1
def removeDublicates_1(nums:list):
    new_list=[]
    for i in nums:
        if i not in new_list:
            new_list.append(i)
    return new_list

print(removeDublicates_1([1,2,3,3,4])) 
print(removeDublicates_1([10, 10, 10]))


# time complexity : O(n)
```

```python
# solution 2

def removeDublicates_2(nums: list):
    for i in nums:
        if nums.count(i)>1:
            nums.remove(i)
    return nums
print(removeDublicates_2([1,2,3,3,4]))
print(removeDublicates_2([10,10,10]))

# time complexity  : O(n)
```

```python
# soltuion 3
# printing unique elements
def removeDublicates_3(nums:list):
    res=1
    n=len(nums)
    for i in range(1,n):
        if(nums[res-1]!=nums[i]):
            nums[res]=nums[i]
            res+=1
    return res

print(removeDublicates_3([1,2,3,3,4]))

# time complexity : O(n)
```
