```python
# prime factorization of a given number
# we need to check, if a number is prime or not

def is_prime(n:int):
    isPrime=True
    if(n==1):
        isPrime=False
    if(n==2 or n==3):
        return isPrime
    if(n%2==0 or n%3==0):
        isPrime=False
    i=5
    while (i*i<=n):
        if(n%i==0 or n%(i+2)==0):
            isPrime=False
        i+=6
    return isPrime

# print(is_prime(65))
# print(is_prime(31))
# time compelxity : O(n^1/2)


# now we need to make a fucntion for prime factorization
def prime_factorization(n:int):
    for i in range(2,n+1):
        if is_prime(i):
            x=i
            while n%x==0:
                print(i, end=" ")
                x*=i

# time complexity : O(n^2)

prime_factorization(100)
# prime_factorization(13)
# prime_factorization(15)
```
