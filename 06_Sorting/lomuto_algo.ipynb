```python
# lomuto's algorithm for array partition
# taking time complexity : O(n)

def lomuto_algorithm(arr:list, p:int):
    n=len(arr)
    arr[p],arr[n-1]=arr[n-1],arr[p]
    pivot=arr[n-1]
    i=-1
    for j in range(0,n):
        if(arr[j]<arr[n-1]):
            i+=1
            arr[i],arr[j]=arr[j],arr[i]
    arr[i+1],arr[n-1]=arr[n-1],arr[i+1]
    return arr

a = [5,13,6,9,12,8,11]
print(lomuto_algorithm(a,5))
```
