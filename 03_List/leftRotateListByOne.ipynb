```python
# left rotate list by one
# [1,2,3] => [2,3,1]

# solution 1
def leftRotate_1(arr:list):
    e=arr.pop(0)
    arr.append(e)
    return arr

print(leftRotate_1([1,2,3]))
```

```python
# solution 2
# time complexity : O(n)
def leftRotate_2(nums:list):
    n=len(nums)
    i=0
    while i<n-1:
        nums[i],nums[i+1]=nums[i+1],nums[i]
        i+=1
    return nums
print(leftRotate_2([1,2,3]))
print(leftRotate_2([10,20,30,40]))
```

```python
# soltuion 3
# time complexity : O(n)
def leftRotate_3(nums:list):
    x=nums[0]
    n=len(nums)
    for i in range(1,n):
        nums[i-1]=nums[i]
    nums[n-1]=x
    return nums

print(leftRotate_3([1,2,3,4]))
```
