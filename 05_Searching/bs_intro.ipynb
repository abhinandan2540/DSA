```python
# binary search
# binary search works well for 'sorted array'

# example 1
# giving a sorted array, we need to return the index value of the input, else -1
# using simplest approach (without binary search)
# this is a linear search
def binary_search_1(arr:list, x:int):
    n=len(arr)
    for i in range(n):
        if arr[i]==x:
            return i
    return -1


my_array = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
print(binary_search_1(my_array,20))
```

```python
# solution 2
def binarySearch_1(arr:list, x:int):
    low=0
    high=len(arr)-1
    while (low<=high):
        mid=(low+high)//2
        if(arr[mid]==x):
            return mid
        elif (arr[mid]>x):
            high=mid-1
        else:
            low=mid+1

    return -1
array = [10,20,30,40,50,60]
print(binarySearch_1(array,25))
```
