```python
# simple linear search

def simpleLinearSearch(arr:list,x:int):
    # simple linear search, using only for loop once
    for i in range(len(arr)):
        if(arr[i]==x):
            return i
    return -1


nums = [0, 20, 80, 30, 60, 50,
        110, 100, 130, 170]
print(simpleLinearSearch(nums,130))
```

```python
# updated version of linear search 
def updatedLinearSearch(arr:list, x:int):
    left=0
    right=len(arr)-1
    res=-1
    for left in range(right):
        if(arr[left]==x):
            res=left
            break
        if(arr[right]==x):
            res=right
            break
        left+=1
        right-=1
    return res


nums = [0, 20, 80, 30, 60, 50,110, 100, 130, 170]
print(updatedLinearSearch(nums,60))
```
