```python
# finding the 1st occurance in a sorted array

def firstOccurance(arr:list,k:int):
    low=0
    high=len(arr)-1
    while(low<=high):
        mid=(low+high)//2
        if(arr[mid]==k):
            while mid>=0:
                if arr[mid]!=k:
                    return mid+1
                mid-=1
            return mid
        elif (arr[mid]>k):
            high=mid-1
        else:
            low=mid+1
    return -1

# print(firstOccurance([1,2,3,4,5,6],5))
print(firstOccurance([1,3,2,2,2,2,2,2,2],2))
```

```python
# another way of doing so
def firstOccurance(arr: list, k: int):
    low = 0
    high = len(arr)-1
    while (low <= high):
        mid = (low+high)//2
        if(arr[mid]>k):
            high=mid-1
        elif (arr[mid]<k):
            low=mid+1
        else:
            if(mid==0 or arr[mid-1]!=arr[mid]):
                return mid
            else:
                high=mid-1
    return -1


print(firstOccurance([1,2,2,2,3,3,3,4], 2))

# time complexity : O(logn)
# space complexity : O(1)
```
