```python
# merge sort algorithm
# combination of merge_subarray problem

def mergeSubarray(arr, low, high, mid):
    arr_1=arr[low:mid+1]
    arr_2=arr[mid+1:high+1]
    m=len(arr_1)
    n=len(arr_2)
    i=0
    j=0
    k=low
    while(i<m and j<n):
        if(arr_1[i]<=arr_2[j]):
            arr[k]=arr_1[i]
            i+=1
            k+=1
        else:
            arr[k]=arr_2[j]
            j+=1
            k+=1
    while(i<m):
        arr[k]=arr_1[i]
        i+=1
        k+=1
    while(j<n):
        arr[k]=arr_2[j]
        j+=1
        k+=1
    return arr


def mergeSort(arr,low, high):
    if (high>low):
        mid=(high+low)//2
        mergeSort(arr,low,mid)
        mergeSort(arr,mid+1,high)
        mergeSubarray(arr,low,high,mid)
    return arr


arr = [10, 5, 30, 15, 7]
print(mergeSort(arr,0,4))

# time complexity : O(nlogn)
# space complexity : O(n)
```
