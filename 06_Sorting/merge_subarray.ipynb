```python
# solution
# array: bifurcate into two arrays, then applying 'merge 2 sort arrays'

def merge_subarray(arr:list, low:int, high:int, mid:int):
    arr_1=arr[low:mid+1]
    arr_2=arr[mid+1:]
    m=len(arr_1)
    n=len(arr_2)
    i=0
    j=0
    res=[]

    while (i<m and j<n):
        if(arr_1[i]<arr_2[j]):
            res.append(arr_1[i])
            i+=1
        if(arr_2[j]<arr_1[i]):
            res.append(arr_2[j])
            j+=1
    while(i<m):
        res.append(arr_1[i])
        i+=1
    while(j<n):
        res.append(arr_2[j])
        j+=1
    return res


l1=[5,8,11,14,7]
print(merge_subarray(l1,0,4,3))
```

```python
# merge two arrays
def merge_two_array(arr_1:list, arr_2:list):
    m=len(arr_1)
    n=len(arr_2)
    res=[]
    i=0
    j=0
    while(i<m and j<n):
        if(arr_1[i]<arr_2[j]):
            res.append(arr_1[i])
            i+=1
        if (arr_1[i]>arr_2[j]):
            res.append(arr_2[j])
            j+=1
    while (i<m):
        res.append(arr_1[i])
        i+=1
    while (j<n):
        res.append(arr_2[j])
        j+=1
    return res

l1=[5,8,11,14]
l2=[7]
print(merge_two_array(l1,l2))
        
```
