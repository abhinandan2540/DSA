```python
# largest element in a list

def largestElement(nums:list):
    max=nums[0]
    if not nums:
        return None
    elif(len(nums)==1):
        return max
    else:
        for i in nums:
            if max<i:
                max,i=i,max
    return max

print(largestElement([10,5,20,8]))
print(largestElement([30,30,8]))
print(largestElement([6]))
   
# time complexity : O(n)   
```

```python
# solution 2

def largestElement_2(nums:list):
    max=nums[0]
    if not nums:
        return None
    else:
        for i in range(1,len(nums)):
            if nums[i]>max:
                max=nums[i]
    return max

print(largestElement_2([10,5,20,8]))
print(largestElement_2([6]))

# time complexity : O(n)
```
