```python
# seperating even and odd values from a list
# solution 1

def seperateEvenOdd(nums:list):
    even=[]
    odd=[]
    for i in nums:
        if(i%2==0):
            even.append(i)
        else:
            odd.append(i)

    return even,odd

l1=[10,41,30,15,80]
l2=[10,30,40]
even,odd=seperateEvenOdd(l1)
print(even)
print(odd)

# time complexity : O(n)
```

```python
# solving this problem with list comprehension

def seperateEvenOdd_2(nums:list):
    even=[i for i in nums if i%2==0]
    odd=[i for i in nums if i%2!=0]
    return even,odd

even,odd = seperateEvenOdd_2([10,41,30,15,80])
print(even)
print(odd)
```
