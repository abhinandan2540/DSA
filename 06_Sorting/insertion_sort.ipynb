insertion sort
It's stable and in-place sorting algorithm
Time complexity O(n^2) in wrost case
in best case O(n)

It works on a hypothesis that, it let, some portion of list is sorted and some unsorted
then it sort the whole list untill it become fully sorted

```python


def insertion(nums:list):
    n=len(nums)
    for i in range(1,n):
        x=nums[i]
        j=i-1
        while(j>=0 and x<nums[j]):
            nums[j+1]=nums[j]
            j-=1
        nums[j+1]=x
    return nums

l=[20,5,40,60,10,30]
print(insertion(l))
```

```python
def insertion_sort(nums:list):
    n=len(nums)
    for i in range(1,n):
        key=nums[i]
        j=i-1
        while(j>=0 and nums[j]>key):
            nums[j+1]=nums[j]
            j-=1
        nums[j+1]=key
    return nums


l1=[8,3,5,2,6]
print(insertion_sort(l1))
```
