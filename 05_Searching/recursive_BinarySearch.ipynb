```python
# binary search
def binarySearch(arr:list, x:int):
    low=0
    high=len(arr)-1
    while(low<=high):
        mid=(low+high)//2
        if(arr[mid]==x):
            return mid
        elif (arr[mid]>x):
            high=mid-1
        else:
            low=mid+1
    return -1


nums = [10, 20, 30, 40, 50, 60]
print(binarySearch(nums,50))
print(binarySearch(nums,25))

# time complexity : O(logn)
```

```python
# making this bianry search into a recursive function
def binarySearchRecursive(arr:list, n:int,high:int, low:int):
    if low>=high:
        return -1
    mid=(high+low)//2
    if(arr[mid]==n):
        return mid
    if(arr[mid]>n):
        high=mid-1
        return binarySearchRecursive(arr,n,high,low)
    else:
        low=mid+1
        return binarySearchRecursive(arr, n, high, low)
    return -1

nums = [10, 20, 30, 40, 50, 60]
high=len(nums)
print(binarySearchRecursive(nums,25,high,0))

# time complexity : O(logn)
```
