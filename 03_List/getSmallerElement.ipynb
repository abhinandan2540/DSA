```python
# get smaller element from a list
# solution 1

def getSmallerElement(nums:list, x:int):
    smlrItems=[]
    for i in nums:
        if(i<x):
            smlrItems.append(i)
    return smlrItems

print(getSmallerElement([8,100,20,40,3,7],10))
print(getSmallerElement([15, 3, 27, 8, 42, 19, 5, 33, 21, 10],30))


# time complexity : O(n)
```

```python
# solving this problem using list comprehension

def getSmallerElement_2(nums:list, n:int):
    l2=[i for i in nums if i<n]
    return l2
print(getSmallerElement_2([8,100,20,40,3,7],10))
```
