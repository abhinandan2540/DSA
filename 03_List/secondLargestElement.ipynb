```python
# finding second largest element
# solution 1
# my solution using sorting algorithm
def secondLargestElement(nums:list):
    n=len(nums)
    for i in range(n-1):
        if nums[i]==nums[i+1]:
            return None
    if len(nums)==1:
        return None
    else:
        for i in range(n-1):
            for j in range(n-1-i):
                if nums[j]>nums[j+1]:
                    nums[j],nums[j+1]=nums[j+1],nums[j]
    return nums[-2]

        
print(secondLargestElement([10,5,8,20]))  
print(secondLargestElement([10,10,10])) 
print(secondLargestElement([10]))
```

```python
# solution 2
def getSecondLargestElement_2(nums:list):
    maxElement=max(nums)
    secMaxElement=None
    if (len(nums)<=1):
        return None
    for i in nums:
        if (i!=maxElement):
            if secMaxElement==None:
                secMaxElement=i
            else:
                secMaxElement=max(i,secMaxElement)
    return secMaxElement

print(getSecondLargestElement_2([10,5,20,4]))


# time complexity : O(n)
```

```python
# solution 3
# efficient approach

def secondLargest(nums:list):
    lar=nums[0]
    slar=None
    if(len(nums)<=1):
        return None
    else:
        for i in nums[1:]:
            if(i>lar):
                slar=lar
                lar=i
            elif i!=lar:
                if slar==None or slar<i:
                    slar=i
    return slar

print(secondLargest([10,20,50,40,30]))
```
