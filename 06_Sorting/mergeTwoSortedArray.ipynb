```python
# merge two sorted arrays

def mergeTwoSortedArray(l1:list, l2:list):
    res=[]
    m=len(l1)
    n=len(l2)
    i=0
    j=0
    while (i<m and j<n):
        if (l1[i]<l2[j]):
            res.append(l1[i])
            i+=1
        else:
            res.append(l2[j])
            j+=1
    
    while(i<m):
        res.append(l1[i])
        i+=1
    while (j<n):
        res.append(l2[j])
        j+=1

    return res


l1=[10,15]
l2=[5,6,6,30,40]
print(mergeTwoSortedArray(l1,l2))

# time complexity : O(m+n)
```
