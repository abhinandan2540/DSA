```python
# efficient approach
# we can use merge sort algorithm

try:

    def count_merge(arr: list, low: list, high: list, mid: list):
        a1 = arr[low:mid+1]
        a2 = arr[mid+1:high+1]
        m = len(a1)
        n = len(a2)
        i = 0
        j = 0
        k = low
        count=0
        while (i < m and j < n):
            if (a1[i] <= a2[j]):
                arr[k] = a1[i]
                i += 1
                # k += 1
            
            else:
                arr[k] = a2[j]
                j += 1
                # k += 1
                count+=(m-i)
            k+=1

        while (i < m):
            arr[k] = a1[i]
            i += 1
            k += 1
        while (j < n):
            arr[k] = a2[j]
            j += 1
            k += 1
        return count


    def count_inversion(arr:list, low:int, high:int):
        count=0
        if(high>low):
            mid=(low+high)//2
            count+=count_inversion(arr,low,mid)
            count+=count_inversion(arr,mid+1,high)
            count+=count_merge(arr,low,high,mid)
            
        return count


    arr = [5, 8, 11, 14, 2, 7]
    print(count_inversion(arr,0,5))

except Exception as e:
    print(e)

# time complexity : O(nlogn)
```
