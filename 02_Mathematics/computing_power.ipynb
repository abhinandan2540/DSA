we have to calcuate power of a certain number

```python

# solution 1
def compute_power(x:int, n:int):
    """ x:base, n:power """
    res=1
    while n>=1:
        res*=x
        n-=1
    return res


compute_power(5,1)

# time complexity : O(n)
```

```python
# solution 2

def compute_power_2(x:int, n:int):
   if(n==0):
      return 1
   return x*compute_power_2(x,n-1)

compute_power_2(4,3)

# time complexity : O(n)
```

```python
# solution 3

def n_power(x,y):
    if(y==0):
        return 1
    temp=n_power(x,y//2)
    if(y%2==0):
        return temp*temp
    else:
        return x*temp*temp
    
print(n_power(4,3))

# time complexity : O(logn)
```
