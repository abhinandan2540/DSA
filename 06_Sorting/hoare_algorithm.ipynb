```python
def hoare_algorithm(arr:list, low:int, high:int):
    """
    arr: array input
    low: starting index of array (eg: 0)
    high: final index of array.
    
    """

    pivot=arr[low]
    i=low-1
    j=high+1

    while True:
        i+=1
        while (arr[i]<pivot):
            i+=1
        j-=1
        while (arr[j]>pivot):
            j-=1
        if(i>=j):
            return j
        arr[i],arr[j]=arr[j],arr[i]


arr = [5, 13, 6, 9, 12, 8, 11]
print(hoare_algorithm(arr, 0, 6))
```
