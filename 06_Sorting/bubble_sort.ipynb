Bubble sort: 
simplest sorting algorithm, it's stable and in-place.

Basic of Bubble sort is: it chooses the large value and put it in the end position of the array and keep doing it 
                         untill it become sorted array.

```python

def bubble_sort(nums: list):
    n = len(nums)
    for i in range(n-1):
        for j in range(n-i-1):
            if (nums[j] > nums[j+1]):
                nums[j], nums[j+1] = nums[j+1], nums[j]
    return nums


l1 = [10,8,20,5]
print(bubble_sort(l1))

# time complexity : O(n^2)
```

```python
# what if your array is sorted from middle

def bubble_sort_2(nums:list):
    nums_sorted=sorted(nums)
    n=len(nums)
    for i in range(n-1):
        for j in range(n-i-1):
            if (nums[j] > nums[j+1]):
                nums[j], nums[j+1] = nums[j+1], nums[j]
        if(nums==nums_sorted):
            return nums
    return nums

l1=[8,5,2,7,9,10]
print(bubble_sort_2(l1))
```

```python
# in another way, we can do it

def bubble_sort_3(nums):
    n=len(nums)
    for i in range(n-1):
        swaped=False
        for j in range(n-i-1):
            if (nums[j] > nums[j+1]):
                nums[j], nums[j+1] = nums[j+1], nums[j]
                swaped=True

        if swaped==False:
            return nums
                
    return nums

l1=[3,5,10,20,40]
print(bubble_sort_3(l1))
    
# for this case
# if 'array' is sorted, then it'll take O(n)
# else, O(n^2)
#  
```
