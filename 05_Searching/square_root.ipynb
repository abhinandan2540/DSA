```python
# square root of a number
import math
def sqr_root(n:int):
    return math.floor(n**0.5)

print(sqr_root(8))
```

```python
def square_root(n:int):
    low=1
    high=n
    ans=-1
    while(low<=high):
        mid=(low+high)//2
        mid_square=mid*mid
        if(mid_square==n):
            return mid
        elif (mid_square>n):
            high=mid-1
        else:
            low=mid+1
            ans=mid
    return ans

print(square_root(26))
```
