```python
# sorting algorithms provide us stability

# example of stability sort
l=[('ayus',50),('aab',50),('puys',60),('zakir',60)]
l.sort()
print(l)

# on the other hand, unstable sort, when sorting happened in not ordered order.
# like sorted output: [('ayus', 50), ('aab', 50).. etc]
```

stable vs unstable sort:

stable sort: in which the relative position of input and output is same
eg: bubble sort, insertion sort, merge sort etc.

Unstable sort: in which, the relative position of input and output is not same.
eg: selection sort, quick sort, heap sort.

in-palce vs not in-place:

in-place: if the algorithm not take any extra space for computation
eg: bubble sort, insertion sort, selection sort, quick sort.

not in-place: take extra space for sorting.
eg: merge sort, counting sort etc.
