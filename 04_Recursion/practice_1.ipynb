```python
# recursion practice
def func(n):
    if n==0:
        return 
    print(n)
    func(n-1)
    print(n)

func(3)
```

```python
def func(n):
    if n==0:
        return 
    func(n-1)
    print(n)
    func(n-1)
func(3)
```

```python
# recursive problem
def func(n):
    if(n<=1):
        return 0
    else:
        return 1+func(n//2)
        
func(16)
```
