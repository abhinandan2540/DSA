quick sort using *hoare's algorithm*

```python
def hoare_algorithm(arr:int, l:int, h:int):
    pivot=arr[l]
    i=l-1
    j=h+1

    while True:
        i+=1
        while(arr[i]<pivot):
            i+=1
        j-=1
        while(arr[j]>pivot):
            j-=1
        if(i>=j):
            return j
        arr[i],arr[j]=arr[j],arr[i]
    

def q_sort(arr: list, l: int, h: int):
    if (l < h):
        p = hoare_algorithm(arr, l, h)
        q_sort(arr, l, p)
        q_sort(arr, p+1, h)
    return arr


a = [5, 13, 6, 9, 12, 8, 11]
print(q_sort(a,0,6))
```
