```python
# twosum without hash

def twoSum(arr, target):
    n=len(arr)
    match=False
    if(n==1):
        return match
    for i in range(n-1):
        for j in range(i+1,n):
            if(arr[i]+arr[j]==target):
                match=True
                return match
    return match

print(twoSum([0,-1,2,-3,1],-2))
```
