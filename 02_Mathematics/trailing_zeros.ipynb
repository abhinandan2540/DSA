```python
# trailing complexity
# concept : number of trailing zeros in a factorial depends on the number of 5's in that number

# number of 0 in a factorial of a number = floor(n/5) + floor(n/25) + floor(n/125) +.....
```

```python
def trailing_zero_1(n:int):
    to=0
    i=5
    while (n//i>=1):
        to+=n//i
        i*=5
    return to
 

print(trailing_zero_1(15))
```
