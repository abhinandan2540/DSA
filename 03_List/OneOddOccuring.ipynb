```python
# one odd occuring
# soltuion 1

# using dictionary
def OneOddOccuring_1(nums:list):
    odd_dict={}
    count=1
    odd_val=0
    for i in range(len(nums)):
        if nums[i] in odd_dict.keys():
            odd_dict[nums[i]]+=count
        else:
            odd_dict[nums[i]]=count
    for k,v in odd_dict.items():
        if (v%2!=0):
            odd_val=k
    return odd_val


print(OneOddOccuring_1([1, 2, 3, 2, 3, 1, 3]))
print(OneOddOccuring_1([5, 7, 2, 7, 5, 2, 5]))
```

```python
# solution 2

# using for loop
def OneOddOccuring_2(nums:list):
    count=0
    for i in nums:
        for j in nums:
            if i==j:
                count+=1
        if count%2!=0:
            return i


print(OneOddOccuring_2([1, 2, 3, 2, 3, 1, 3]))
print(OneOddOccuring_2([5, 7, 2, 7, 5, 2, 5]))
```

```python
# solution 3
# using xor
# xor return odd value

def OneOddOccuring_3(nums:list):
    odd_val=0
    for i in nums:
        odd_val=odd_val^i
    return odd_val


print(OneOddOccuring_3([1, 2, 3, 2, 3, 1, 3]))
print(OneOddOccuring_3([5, 7, 2, 7, 5, 2, 5]))
```
