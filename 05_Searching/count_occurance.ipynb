```python
# counting occurance in a sorted array
# using linear search
def countOccurance(nums:list, x:int):
    count=0
    for i in nums:
        if i==x:
            count+=1
    return count
arr=[10,20,20,20,30,30]
print(countOccurance(arr,20))
```
