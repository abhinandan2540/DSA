```python
def intersection(a:list, b:list):
    res=[]
    m=len(a)
    n=len(b)
    if(m>n):
        for v in a:
            if v in b:
                res.append(v)
    else:
        for v in b:
            if v in a:
                res.append(v)

    return list(set(res))


a=[1,1,3,3]
b=[1,1,3,5,7]
print(intersection(a,b))


# time complexity : O(m.n)
```

```python
# intersection 2

def intersection(a:list, b:list):
        i=0
        j=0
        res=[]
        m=len(a)
        n=len(b)
        while(i<m and j<n):
            if(i>0 and a[i]==a[i-1]):
                i+=1
                continue

            if(a[i]<b[j]):
                i+=1

            if(a[i]>b[j]):
                j+=1
                
            if(a[i]==b[j]):
                res.append(a[i])
                i+=1
                j+=1

        return res

a=[1,1,3,3]
b=[1,1,3,5,7]
print(intersection(a,b))

# time complexity : O(m+n)
```
