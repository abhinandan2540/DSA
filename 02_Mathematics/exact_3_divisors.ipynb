```python
# exact 3 divisors
# naive approach
def n_divisors(n:int):
    count=0
    for i in range(1,n+1):
        if(n%i==0):
            count+=1
    return count

def exact3Divisors(n:int):
    exact3div_count=0
    for i in range(1,n+1):
        if(n_divisors(i)==3):
            exact3div_count+=1
    return exact3div_count


print(exact3Divisors(10))
```

```python
# method 2

# square of prime number is only divisible by 1, pn^2 and pn.
# so, according to this thing
# it gives us exactly  3 divisors

def is_prime(n:int):
    isPrime=True
    if (n==1):
        isPrime=False
    if (n==2 or n==3):
        return isPrime
    if (n%2==0 or n%3==0):
        isPrime=False
    i=5
    while (i*i<=n):
        if(n%i==0) or (n%(i+2)==0):
            isPrime=False
        i+=6
    return isPrime
    
def exact3Divisors_1(n:int):
    count=0
    i=2
    while(i*i<=n):
        if(is_prime(i)):
            count+=1
        i+=1
    return count

print(exact3Divisors_1(30))

# time complexity : O(n^1/2*n^1/4)
# or O(n^1/2)
```
