Finding the LCM of two numbers

```python

def lcm_1(a:int, b:int):
    res=max(a,b)
    while True:
        if (res%a==0 and res%b==0):
            return res
        res+=1
    
print(lcm_1(4,6))
print(lcm_1(7,3))

# in this code
# time complexity : O(a*b)
# space complexity : O(1)
```

```python
# another way
# we know the formula
# hcf*lcm=a*b
# lcm=a*b/(hcf)

def gcd(a:int, b:int):
    if (b==0):
        return a
    return gcd(b, a%b)

def lcm_2(a:int,b:int):
    return a*b//(gcd(a,b))

print(lcm_2(4,6))
print(lcm_2(7,3))

# time complexity : O(log(min(a,b)))
```
