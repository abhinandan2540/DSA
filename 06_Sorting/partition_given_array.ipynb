```python
# partition given array
arr=[5,13,6,9,12,8,11]
print(arr)

# let say we want every samll element (than 8) in one side, and every other in another

def naive_partition(arr, p):
    # p=5 #index of 8
    n=len(arr)
    arr[p],arr[n-1]=arr[n-1],arr[p]
    res=[]
    for i in range(n):
        if arr[i]<=arr[n-1]:
            res.append(arr[i])
    for i in range(n):
        if(arr[i]>arr[n-1]):
            res.append(arr[i])

    return res

print(naive_partition(arr,5))
# time complexity : O(n)
# space complexity : O(n)
```

```python
# what can be done to make space complexity : O(n)
# lomuto algorithm for partition

def lomuto_algorithm(arr:list, p:int):
    """ p: partition number index """
    n=len(arr)
    # swapping partition number
    arr[p],arr[n-1]=arr[n-1],arr[p]
    pivot=arr[n-1]
    l=0
    i=l-1
    for j in range(l,n):
        if(arr[j]<pivot):
            i+=1
            arr[i],arr[j]=arr[j],arr[i]
    arr[i+1],arr[n-1]=arr[n-1],arr[i+1]
    return arr

arr=[5,13,6,9,12,8,11]
print(lomuto_algorithm(arr,5))

# time complexity : O(n)
# auxilarry space complexity : O(1)
```

```python
# another efficient method is hoare's algorithm
# which is more efficient and faster than lomuto's algorithm

def hoare_algorithm(arr:list, low:int, high:int):
    pivot=arr[low]
    i=low-1
    j=high+1
    while True:
        i+=1
        while(arr[i]<pivot):
            i+=1
        j-=1
        while(arr[j]>pivot):
            j-=1
        if(i>=j):
            return j
        arr[i],arr[j]=arr[j],arr[i]
    
  

arr = [5,13,6,9,12,8,11]
print(hoare_algorithm(arr,0,6))

```
