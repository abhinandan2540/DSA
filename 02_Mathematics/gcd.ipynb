```python
# GCD finding using euclidean algorithm

def gcd_1(a:int, b:int):
    while a!=b:
        if (a>b):
            a-=b
        else:
            b-=a
    return a

print(gcd_1(4,6))
print(gcd_1(7,13))

# this algorithm has 
# time complexity : O(min(a,b)) or O(n)
# space complexity : O(1)
```

```python
# making a more optimized solution

def gcd_2(a:int, b:int):
    if (b==0):
        return a
    return gcd_2(b, a%b)

print(gcd_2(4,6))
print(gcd_2(100,200))

# in this problem
# time complexity : O(log(min(a,b)))
# space complexity : O(1)
```
