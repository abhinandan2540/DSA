```python
# finding list mean

l1=[10,20,30,40,50]
l2=[30,60,40]
def list_mean(num_list:list):
    n=len(num_list)
    n_sum=0
    for i in num_list:
        n_sum+=i
    return (n_sum/n)

print(list_mean(l2))
```
