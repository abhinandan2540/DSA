we have to calcuate power of a certain number

```python

# solution 1
def compute_power(x:int, n:int):
    """ x:base, n:power """
    res=1
    while n>=1:
        res*=x
        n-=1
    return res


compute_power(5,1)

# time complexity : O(n)
```

```python
# solution 2

def compute_power_2(x:int, n:int):
    res=1
    if(n>0):
        res*=x
        print(x)
    return compute_power_2(x,n-1)

compute_power_2(2,3)
```
