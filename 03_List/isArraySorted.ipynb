```python
# for a given array, we need to check, if the given array is sorted or not in non-decreasing order (increasing order).

def isSorted(nums:list):
    n=len(nums)
    sorted=True
    if (n<=1):
        return sorted
    else:
        for i in range(n-1):
            if(nums[i]>=nums[i+1]):
                sorted=False
            else:
                return sorted
    return sorted


def isArraySorted(nums:list):
    if isSorted(nums):
        print('YES')
    else:
        print('NO')



isArraySorted([10,20,20,30,40,50])
isArraySorted([10])
isArraySorted([])
```

```python
# we can do it using while loop too

def isSorted(nums:list):
    sorted=True
    n=len(nums)
    i=0
    if(n<=1):
        return sorted
    while i>n:
        if nums[i]>nums[i+1]:
            sorted=False
            break
        i+=1
    return sorted

def isArraySorted(nums:list):
    if isSorted(nums):
        print('YES')
    else:
        print('NO')

isArraySorted([10, 20, 20, 30, 40, 50])
isArraySorted([10])
```

```python
# another using while loop

def isSorted(nums:list):
    if(len(nums)<=1):
        return True
    i=1
    while i<len(nums):
        if nums[i]<nums[i-1]:
            return False
        i+=1
    return True
        
def isArraySorted(nums:list):
    if isSorted(nums):
        print('YES')
    else:
        print('NO')

isArraySorted([10,20,20,30,40])
```

```python
# the difference between sort() and sorted()
# sort() sort the main list, sorted() create another list of sorted version from the first list

l1=[20,30,10,5,4]
l1.sort()
print(l1)
l2=sorted(l1)
print(l2)

# same concept with reverse(), reversed()
r1=[20,30,10,5,4]
r1.reverse()
print(r1)
r2=reversed(r1)
print(r2)
```

```python
# another method

def isSorted(nums:list):
    l2=sorted(nums)
    if nums==l2:
        return True
    else:
        return False
    
def isArraySorted(nums:list):
    if isSorted(nums):
        print('YES')
    else:
        print('NO')

isArraySorted([10, 20,20,30,40])
isArraySorted([30,20,10,50])
```

```python
# another best method of doing this
# 1 : sorted array (increasing or decreasing order)
# 0 : unsorted array

def isSorted(nums:list):
    incFlag=True
    decFlag=True
    n=len(nums)
    if(n<=1):
        return 1
    for i in range(n-1):
        if (nums[i]>nums[i+1]):
            incFlag=False
        if (nums[i]<nums[i+1]):
            decFlag=False
        if not incFlag and not decFlag:
            return 0
    return 1
print(isSorted([10,20,30]))
print(isSorted([30,20,10]))
print(isSorted([30,10,20]))
```
